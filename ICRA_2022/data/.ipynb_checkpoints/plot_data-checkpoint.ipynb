```python
import pandas as pd
import numpy as np
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure
output_notebook()
```

```python
def plot_data(data, thresh, start_ind, plot_title, plot_name):
    opts = dict(width=800, height=250, min_border=0)
    colors = ['blue','red','green','magenta']
    
    # create a Figure object
    p = figure(width=900, height=300, tools="pan,reset,save", x_axis_label='Time (s)',  y_axis_label='Signal', title=plot_title)

    # gather data to plot
    xs = []
    ys = []
    for i in range(len(data.columns)-1):
        xs.append(data.Time[start_ind:] - data.Time[start_ind])
        ys.append(data["Sensor"+str(i+1)][start_ind:])
    
    # plot lines
    p.multi_line(xs, ys, line_color=colors[:len(data.columns)], line_width=2)
    
    # plot threshold
    p.line(data.Time[start_ind:] - data.Time[start_ind], thresh*np.ones((len(data[start_ind:]),)), line_color='black', line_dash = 'dotted', line_width=2)

    # specify how to output the plot(s)
#     output_file(plot_name + ".jpg")

    # display the figure
    show(p)
    
def process_data(task, file_name, thresh, start_ind, plot_title, plot_name):
    if task == "egg":
        task_str = "egg_hist_"
        sens_num = 4
    elif task == "wing_nut":
        task_str = "hist_"
        sens_num = 1
    elif task == "scissors":
        task_str = "purple_scissor_hist_"
        sens_num = 4

#     file_name = task_str + topic + "_" + number + ".csv"
    cols = ['Time']
    for i in range(sens_num):
        cols.append('Sensor'+str(i+1))

    data = pd.read_csv(file_name+".csv", names=cols)
    print(data.head())

    plot_data(data, thresh, start_ind, plot_title,plot_name)
```

```python
# task: "wing_nut", "egg", "scissors", "bottle"
# topic: "pres", "force"
# number: '0912131215'

# process_data("egg", "pres", "0912131054", 20, 5, "Egg Manipulator Sensor Readings","egg")
process_data("wing_nut", "wing_hist_pres_0915193624", 200, 51, "Wing Nut Manipulator Sensor Readings","egg")
```
